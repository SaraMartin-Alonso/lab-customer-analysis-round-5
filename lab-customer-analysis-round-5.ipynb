```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
# X-y split
target = data['Total Claim Amount']
target.head()
```

```python
data_X = data.drop(['Total Claim Amount'], axis=1)
data_X
```

```python
# Separate numerical from categorical
X_num = data_X.select_dtypes(include = np.number)
X_cat = data_X.select_dtypes(include = np.object)
```

```python
#X_numerical --> scale data
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
 
transformer = Normalizer().fit(X_num)
x_normalized = transformer.transform(X_num)
 
X_num=pd.DataFrame(x_normalized)
X_num.head()
```

```python
X_cat=X_cat.drop(['Effective To Date'], axis=1)
X_cat=X_cat.drop(['Customer'], axis=1)
X_cat=X_cat.drop(['State'], axis=1)
X_cat=X_cat.drop(['Policy'], axis=1)
X_cat=X_cat.drop(['Policy Type'], axis=1)
X_cat=X_cat.drop(['Coverage'], axis=1)
X_cat=X_cat.drop(['Renew Offer Type'], axis=1)

# for i in range(len(X_cat)):
#    X_cat[X_cat.columns[i]].unique()
```

```python
X_cat
```

```python
# One Hot Encode X_cat
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='error').fit(X_cat)
X_cat = encoder.transform(X_cat).toarray()
X_cat = pd.DataFrame(X_cat)


```

```python
X_cat.head()
```

```python
# Concatenate numerical and categorical
data_concat = pd.concat([X_num, X_cat], axis=0)
data_concat
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_concat, target, test_size=0.4, random_state=100)
```
